**Question:** what does the 5.397605e-79 value mean? Does it represent the value 5 or a missing value?

We can answer this question by 
- compiling all the data to a question (represented by a column),
- getting the count of each value in the data,
- and comparing it to the count [here](https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/DPQ_G.htm#Data_Processing_and_Editing).

```python
# compile data
# let's check with DPQ010
mental_health_df.groupby(['DPQ010']).size()
```

```python
# let's check it again with a different way because it never hurts to be extra cautious
mental_health_df['DPQ010'].value_counts()
```

Oh no! The count doesn't match up! This is because the data I'm using is from 2011-2016 while the data the chart has is from 2011-2012. Let's add up all the counts from 2011 to 2016 manually since I don't want to do it the other way.

**0:** 3726 + 3976 + 3805 = 11507 \
**1:** 800 + 881 + 852 = 2533 \
**2:** 214 + 293 + 255 = 762 \
**3:** 201 + 242 + 239 = 682 \
**7:** 2 + 1 + 7 = 10 \
**9:** 8 + 5 + 6 = 19 \
**.:** 664 + 526 + 571 = 1761

We have our **answer**: the 5.397605e-79 value represents a 0.
